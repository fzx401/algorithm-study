```python
from gurobipy import *
import scipy as sp
import pandas as pd
```

```python
m = Model()

x = m.addVars(6, 7, lb=0)
y = m.addVars(6, lb=0)

m.addConstr(quicksum(x[i, 6] for i in range(6)) + y[1] >= 3)
m.addConstr(x[5, 3] + y[5] >= 3)
A = m.getA()
denseMatrix = pd.DataFrame(data=sp.sparse.csr_matrix.todense(A))
print(denseMatrix.T)
```
